```python
def fill_missing(arr, i):
    while np.isnan(arr[i]):
        i-=1
    return arr[i]
```

```python
years = range(1970,2015)
us_expect =  [us_expect[i] if np.isnan(us_expect[i])==False else fill_missing(us_expect,i) for i in range(1,len(us_expect))]
us_expect = [us_expect[0]] + us_expect

india_expect =  [india_expect[i] if np.isnan(india_expect[i])==False else fill_missing(india_expect, i) for i in range(1,len(india_expect))]
india_expect = [india_expect[0]] + india_expect

china_expect =  [china_expect[i] if np.isnan(china_expect[i])==False else fill_missing(china_expect,i) for i in range(1,len(china_expect))]
china_expect = [china_expect[0]] + china_expect
print(china_expect)

plt.plot(years, brit_expect,label='Britain')
plt.plot(years,india_expect,label='India')
plt.plot(years,china_expect,label='China')
plt.plot(years,us_expect,label='United States')
plt.title('Tertiary Education Life Expectancy')
plt.xlabel('Year')
plt.ylabel('Life Expectancy (Years)')
plt.legend(loc='upper left')
```
