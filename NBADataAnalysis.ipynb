```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.dates import date2num
import datetime
```

```python
def year(col):
    if (eval(col[:col.find('/',0)]) < 7):
        return eval(col[(col.find('/',3) + 1):]) - 1
    return eval(col[(col.find('/',3) + 1):])

all_data = pd.DataFrame()
for i in range(5, 18): 
    reg_csv = str(i) + '.csv'
    playoff_csv = str(i) + 'Playoff.csv'
    if i < 10:
        reg_csv = '0' + str(i) + '.csv'
        playoff_csv = '0' + str(i) + 'Playoff.csv'
    reg = pd.read_csv(reg_csv)
    playoff = pd.read_csv(playoff_csv)
    playoff['G'] = playoff['G'].apply(lambda x:x+82)
    all_data = all_data.append(reg.append(playoff))
    

all_data['Year'] = all_data['Date'].apply(year)


```

```python
all_data = all_data.drop(['Age', 'GS', 'Date'], axis = 1)
```

```python
all_data = all_data[(all_data['PTS'] != 'Did Not Play') & (all_data['PTS'] != 'Did Not Dress') & (all_data['PTS'] != 'Inactive') & (all_data['PTS'] != 'Not With Team')]
```

```python
def str_to_int(x):
    if type(x) is str:
        return eval(x)
    return x
cols_list = ['FG', 'FGA', 'FG%', '3P', '3PA', '3P%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'GmSc', '+/-']

for i in cols_list:
    all_data[i] = all_data[i].apply(str_to_int)

all_data['Points Shot Attempt'] = all_data['PTS']/all_data['FGA']
all_data['2Pt%'] = (all_data['FG'] - all_data['3P'])/all_data['FG']
```

```python
five = all_data[(all_data['Year'] == 2005) & (all_data['G'] < 83)]
six = all_data[(all_data['Year'] == 2006) & (all_data['G'] < 83)]
seven = all_data[(all_data['Year'] == 2007) & (all_data['G'] < 83)]
eight = all_data[(all_data['Year'] == 2008) & (all_data['G'] < 83)]
nine = all_data[(all_data['Year'] == 2009) & (all_data['G'] < 83)]
ten = all_data[(all_data['Year'] == 2010) & (all_data['G'] < 83)]
eleven = all_data[(all_data['Year'] == 2011) & (all_data['G'] < 83)]
twelve = all_data[(all_data['Year'] == 2012) & (all_data['G'] < 83)]
thirteen = all_data[(all_data['Year'] == 2013) & (all_data['G'] < 83)]
fourteen = all_data[(all_data['Year'] == 2014) & (all_data['G'] < 83)]
fifteen = all_data[(all_data['Year'] == 2015) & (all_data['G'] < 83)]
sixteen = all_data[(all_data['Year'] == 2016) & (all_data['G'] < 83)]
seventeen = all_data[(all_data['Year'] == 2017) & (all_data['G'] < 83)]
five_p = all_data[(all_data['Year'] == 2005) & (all_data['G'] > 82)]
six_p = all_data[(all_data['Year'] == 2006) & (all_data['G'] > 82)]
seven_p = all_data[(all_data['Year'] == 2007) & (all_data['G'] > 82)]
eight_p = all_data[(all_data['Year'] == 2008) & (all_data['G'] > 82)]
nine_p = all_data[(all_data['Year'] == 2009) & (all_data['G'] > 82)]
ten_p = all_data[(all_data['Year'] == 2010) & (all_data['G'] > 82)]
eleven_p = all_data[(all_data['Year'] == 2011) & (all_data['G'] > 82)]
twelve_p = all_data[(all_data['Year'] == 2012) & (all_data['G'] > 82)]
thirteen_p = all_data[(all_data['Year'] == 2013) & (all_data['G'] > 82)]
fourteen_p = all_data[(all_data['Year'] == 2014) & (all_data['G'] > 82)]
fifteen_p = all_data[(all_data['Year'] == 2015) & (all_data['G'] > 82)]
sixteen_p = all_data[(all_data['Year'] == 2016) & (all_data['G'] > 82)]
seventeen_p = all_data[(all_data['Year'] == 2017) & (all_data['G'] > 82)]

year = ['2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']
reg_season = [five, six, seven, eight, nine, ten, eleven, twelve, thirteen, fourteen, fifteen, sixteen, seventeen]
playoffs = [five_p, six_p, seven_p, eight_p, nine_p, ten_p, eleven_p, twelve_p, thirteen_p, fourteen_p, fifteen_p, sixteen_p, seventeen_p]
```

```python
reg_td = []
playoff_td = []
reg_plus_minus = []
playoff_plus_minus = []
for i,j in zip(reg_season, playoffs):
    reg_td_data = i[(i['PTS'] >= 10) & (i['AST'] >= 10) & (i['TRB'] >= 10)]
    playoff_td_data = j[(j['PTS'] >= 10) & (j['AST'] >= 10) & (j['TRB'] >= 10)]
    reg_plus_minus += [sum(i['+/-'])/len(i)]
    playoff_plus_minus += [sum(j['+/-'])/len(j)]
    reg_td.append(len(reg_td_data)/len(i))
    playoff_td.append(len(playoff_td_data)/len(j))

print(reg_plus_minus,playoff_plus_minus)
plt.figure(figsize=(8, 3))
plt.bar(year, reg_td, color='b')
plt.bar(year, playoff_td, color='g')


plt.title('Triple Double Rate Regular Season vs Playoffs')
plt.xlabel('Year')
plt.ylabel('Triple Doubles Rate')
```

```python
data = [reg_plus_minus,playoff_plus_minus]
plt.boxplot(data,labels=['Regular Season','Playoffs'])
plt.title('Lebron James +/- in Regular Season vs Playoffs')
plt.xlabel('Regular Season vs. Playoffs')
plt.ylabel('+/-')
```

```python
playoff_pts = []
reg_pts = []
reg_efficiency = []
playoff_efficiency = []
reg_2_percent = []
playoff_2_percent = []
for i,j in zip(reg_season, playoffs):
    reg_pts.append(i['PTS'].mean())
    playoff_pts.append(j['PTS'].mean())
    reg_efficiency.append(i['Points Shot Attempt'].mean())
    playoff_efficiency.append(j['Points Shot Attempt'].mean())
    reg_2_percent.append(i['2Pt%'].mean())
    playoff_2_percent.append(j['2Pt%'].mean())

plt.figure(figsize=(8, 3))

plt.plot(year, reg_pts, color='b', label='Regular Season')
plt.plot(year, playoff_pts, color='g', label='Playoffs')
plt.title('Points Per Game Regular Season vs Playoffs')
plt.legend(loc='upper right')
plt.ylabel('Year')
plt.xlabel('PPG')
```

```python


psa = pd.DataFrame(columns=['reg/play']+year)
psa.loc[len(psa)] = ['Regular Season'] + reg_efficiency
psa.loc[len(psa)] = ['Playoffs'] + playoff_efficiency


plt.figure(figsize=(7, 3))
plt.plot(year, reg_efficiency, color='b', label='Regular Season')
plt.plot(year, playoff_efficiency, color='g', label='Playoffs')
plt.title('Points per Shot Attempt Regular Season vs Playoffs')
plt.legend(loc='bottom right')
plt.ylabel('Year')
plt.xlabel('Points per Shot Attempt')
```
